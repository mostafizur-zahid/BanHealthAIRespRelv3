# Task
Perform a language character analysis on the 'Symptoms Genre', 'Used Prompt', and 'Response' columns by defining functions to classify individual characters as Bangla, English, or neither, calculating the percentages of these character types for each column, and displaying the results.

## Define Character Classification Functions

### Subtask:
Create Python functions to classify individual characters as either Bangla, English, or neither, based on their Unicode ranges.


**Reasoning**:
To classify individual characters as either Bangla or English, I will define two functions: `is_bangla_char` which checks Unicode ranges for Bangla characters and `is_english_char` which checks for ASCII English letters.



```python
def is_bangla_char(char):
    """
    Checks if a character is a Bangla character based on its Unicode range.
    Bangla script Unicode range: U+0980 to U+09FF
    """
    return '\u0980' <= char <= '\u09FF'

def is_english_char(char):
    """
    Checks if a character is an English letter (uppercase or lowercase).
    """
    return 'A' <= char <= 'Z' or 'a' <= char <= 'z'

print("Functions `is_bangla_char` and `is_english_char` defined.")
```

## Analyze Character Percentages per Column

### Subtask:
Apply the character classification functions to each entry in the 'Symptoms Genre', 'Used Prompt', and 'Response' columns to count Bangla, English, and other characters, and then calculate their respective percentages for each column.


**Reasoning**:
To analyze the language composition, I need to define a function that iterates through a string and counts Bangla, English, and other characters using the previously defined helper functions. This function will be applied to each entry in the target columns.



```python
def analyze_language_composition(text):
    if not isinstance(text, str):
        text = str(text) # Convert non-string types to string
    total_chars = len(text)
    bangla_chars = 0
    english_chars = 0
    other_chars = 0

    for char in text:
        if is_bangla_char(char):
            bangla_chars += 1
        elif is_english_char(char):
            english_chars += 1
        else:
            other_chars += 1

    return pd.Series([total_chars, bangla_chars, english_chars, other_chars])

print("Function `analyze_language_composition` defined.")
```

**Reasoning**:
Now that the `analyze_language_composition` function is defined, I will apply it to the 'Symptoms Genre', 'Used Prompt', and 'Response' columns of the DataFrame `df` to get the character counts, and then calculate the percentage of Bangla, English, and other characters for each column.



```python
symptoms_genre_composition = df['Symptoms Genre'].apply(analyze_language_composition)
symptoms_genre_composition.columns = ['total_chars', 'bangla_chars', 'english_chars', 'other_chars']

used_prompt_composition = df['Used Prompt'].apply(analyze_language_composition)
used_prompt_composition.columns = ['total_chars', 'bangla_chars', 'english_chars', 'other_chars']

response_composition = df['Response'].apply(analyze_language_composition)
response_composition.columns = ['total_chars', 'bangla_chars', 'english_chars', 'other_chars']

# Calculate percentages for 'Symptoms Genre'
total_symptoms_genre_chars = symptoms_genre_composition['total_chars'].sum()
percent_bangla_symptoms_genre = (symptoms_genre_composition['bangla_chars'].sum() / total_symptoms_genre_chars) * 100
percent_english_symptoms_genre = (symptoms_genre_composition['english_chars'].sum() / total_symptoms_genre_chars) * 100
percent_other_symptoms_genre = (symptoms_genre_composition['other_chars'].sum() / total_symptoms_genre_chars) * 100

print("\nLanguage Composition for 'Symptoms Genre':")
print(f"  Bangla Characters: {percent_bangla_symptoms_genre:.2f}%")
print(f"  English Characters: {percent_english_symptoms_genre:.2f}%")
print(f"  Other Characters: {percent_other_symptoms_genre:.2f}%")

# Calculate percentages for 'Used Prompt'
total_used_prompt_chars = used_prompt_composition['total_chars'].sum()
percent_bangla_used_prompt = (used_prompt_composition['bangla_chars'].sum() / total_used_prompt_chars) * 100
percent_english_used_prompt = (used_prompt_composition['english_chars'].sum() / total_used_prompt_chars) * 100
percent_other_used_prompt = (used_prompt_composition['other_chars'].sum() / total_used_prompt_chars) * 100

print("\nLanguage Composition for 'Used Prompt':")
print(f"  Bangla Characters: {percent_bangla_used_prompt:.2f}%")
print(f"  English Characters: {percent_english_used_prompt:.2f}%")
print(f"  Other Characters: {percent_other_used_prompt:.2f}%")

# Calculate percentages for 'Response'
total_response_chars = response_composition['total_chars'].sum()
percent_bangla_response = (response_composition['bangla_chars'].sum() / total_response_chars) * 100
percent_english_response = (response_composition['english_chars'].sum() / total_response_chars) * 100
percent_other_response = (response_composition['other_chars'].sum() / total_response_chars) * 100

print("\nLanguage Composition for 'Response':")
print(f"  Bangla Characters: {percent_bangla_response:.2f}%")
print(f"  English Characters: {percent_english_response:.2f}%")
print(f"  Other Characters: {percent_other_response:.2f}%")
```

# Task
Create a Python function to convert English numerals (0-9) within a string to their corresponding Bangla numerals.

## Define Character Conversion and Cleaning Functions

### Subtask:
Create a Python function to convert English numerals (0-9) within a string to their corresponding Bangla numerals.


**Reasoning**:
To convert English numerals to Bangla numerals, I will define a Python function `english_to_bangla_numerals` that iterates through the input string and replaces English digits with their Bangla equivalents using a mapping dictionary.



```python
def english_to_bangla_numerals(text):
    """
    Converts English numerals (0-9) within a string to their corresponding Bangla numerals.
    """
    bangla_numerals_map = {
        '0': '০', '1': '১', '2': '২', '3': '৩', '4': '৪',
        '5': '৫', '6': '৬', '7': '৭', '8': '৮', '9': '৯'
    }

    converted_text = []
    for char in text:
        if char in bangla_numerals_map:
            converted_text.append(bangla_numerals_map[char])
        else:
            converted_text.append(char)

    return ''.join(converted_text)

print("Function `english_to_bangla_numerals` defined.")
```

## Define Character Conversion and Cleaning Functions

### Subtask:
Create a Python function to clean text by separating Bangla punctuation, filtering for only Bangla letters and numerals, and normalizing whitespace.


**Reasoning**:
To clean the text as per the subtask, I need to define a Python function `clean_bangla_text` that performs punctuation separation, character filtering (keeping only Bangla letters, numerals, and specified punctuation), and whitespace normalization. This function will use regular expressions and character range checks.



**Reasoning**:
The previous code failed due to a `SyntaxError: unexpected character after line continuation character`. This typically happens when the backslash `\` is misused for line continuation. To fix this, I will enclose the multi-line `if` condition in parentheses, which allows for implicit line continuation without the need for explicit backslashes, making the code more robust and readable.



```python
import re

def clean_bangla_text(text):
    """
    Cleans text by separating common punctuation, filtering for only Bangla letters and numerals,
    and normalizing whitespace.
    """
    if not isinstance(text, str):
        text = str(text)

    # 1. Punctuation separation: Add space before and after common punctuation
    # This helps in tokenization by ensuring punctuation is treated as separate tokens
    text = re.sub(r'([.,!?;])', r' \1 ', text)

    # 2. Character filtering: Keep only Bangla letters, Bangla numerals, and common punctuation/spaces
    # Bangla letters Unicode range: U+0980 to U+09FF
    # Bangla numerals Unicode range: U+09E6 to U+09EF
    # Common punctuation: space, comma, period, question mark, exclamation mark, semicolon
    cleaned_chars = []
    for char in text:
        if (('ঀ' <= char <= '৿') or
            ('০' <= char <= '৯') or
            char in [' ', ',', '.', '?', '!', ';']):
            cleaned_chars.append(char)
    filtered_text = ''.join(cleaned_chars)

    # 3. Normalize whitespace: Replace multiple spaces with a single space and remove leading/trailing spaces
    filtered_text = re.sub(r'\s+', ' ', filtered_text).strip()

    return filtered_text

print("Function `clean_bangla_text` defined.")
```
